WEB CRAWLING

Link chosen on INDIA: -https://en.wikipedia.org/wiki/India

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re
```

```python
url = urlopen("https://en.wikipedia.org/wiki/India")
bsobj=soup(url.read())
```

```python
url_process_stack = []
url_process_stack.append(bsobj)
```

EXTRA LINKS IN THE WEBPAGE

```python
for link in bsobj.findAll('a'):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

```python
bsobj.findAll('a')
```

```python
bsobj.find('div',{'id':'bodyContent'})
```

```python
for link in bsobj.find('div',{'id':'bodyContent'}).findAll('a',href = re.compile('^(/wiki/)((?!:).*$)')):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```
